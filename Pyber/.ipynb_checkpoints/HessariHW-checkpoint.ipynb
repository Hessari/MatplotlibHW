```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "/Users/hessari/Desktop/MatplotlibHW/Pyber/data/city_data.csv"
ride_data_to_load = "/Users/hessari/Desktop/MatplotlibHW/Pyber/data/ride_data.csv"

# Read the City and Ride Data

city_data_pd = pd.read_csv(city_data_to_load)
ride_data_pd = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset

combined_city_ride_data = pd.merge(city_data_pd, ride_data_pd, how='outer', on='city')

# Display the data table for preview

combined_city_ride_data.head()
```

```python
cities_avg_fare = combined_city_ride_data.groupby("city").mean()["fare"]
cities_total_rides = combined_city_ride_data.groupby("city").count()["ride_id"] 
cities_types = combined_city_ride_data.loc[:, ["city", "type", "driver_count"]]
cities_types = cities_types.drop_duplicates()
cities_df = cities_types.set_index("city")
cities_df["Average Fare"] = cities_avg_fare
cities_df["Total Rides"] = cities_total_rides
cities_df = cities_df.rename(columns={"type": "City Type", "driver_count": "Total Drivers"})
cities_df = cities_df[["Average Fare", "Total Rides", "Total Drivers", "City Type"]]
cities_df.head()
```

```python
# Axes    
x_axis = cities_df["Total Rides"]
y_axis = cities_df["Average Fare"]
z_axis = cities_df["Total Drivers"]
colors = ["gold", "lightskyblue", "lightcoral"]
# Plot configurations
plt.title("Pyber Ride Sharing Data (2016)")
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare($)")
plt.xlim(0, 41)
plt.ylim(17, 45)
plt.grid(True)
# Plot
plt.scatter(x_axis, y_axis, s=z_axis*17, alpha=0.7, linewidth=1.1, color="gold", edgecolor="black")
plt.show()
```
